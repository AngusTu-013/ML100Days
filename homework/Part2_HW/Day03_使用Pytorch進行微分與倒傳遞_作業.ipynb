é€™è£¡æˆ‘å€‘å¾ˆç°¡å–®çš„å¯¦ä½œå…©å±¤çš„ç¥ç¶“ç¶²è·¯é€²è¡Œå›æ­¸å•é¡Œï¼Œå…¶ä¸­loss functionç‚ºL2 loss

$$
L2\_loss = (y_{pred}-y)^2
$$

å…©å±¤ç¶“ç¶²è·¯å¦‚ä¸‹æ‰€ç¤º
$$
y_{pred} = ReLU(XW_1)W_2

```python
import torch
device = torch.device('cpu')
```

```python
# N: batch size
# D_in: input dimension
# H: hidden dimension
# D_out: output dimension
N, D_in, H, D_out = 64, 1000, 100, 10

# éš¨æ©Ÿç”Ÿæˆx, y
###<your code>###
x = torch.randn((N, D_in)).to(device)
y = torch.randn((N, D_out)).to(device)

# åˆå§‹åŒ–weight W1, W2
###<your code>###
W1 = torch.randn((D_in, H), requires_grad=True).to(device)
W2 = torch.randn((H, D_out), requires_grad=True).to(device)

# è¨­ç½®learning rate
learning_rate = 1e-6

# è¨“ç·´500å€‹epoch
for t in range(500):
    # å‘å‰å‚³é: è¨ˆç®—y_pred
    ###<your code>###
    # ğ‘¦_ğ‘ğ‘Ÿğ‘’ğ‘‘ =ğ‘…ğ‘’ğ¿ğ‘ˆ(ğ‘‹ğ‘Š1)ğ‘Š2
    h = torch.matmul(x, W1)
    h_relu = torch.relu(h)
    y_pred = torch.relu(h).mm(W2)

    # è¨ˆç®—loss
    ###<your code>### 
    loss = ((y_pred - y)**2).sum()    #ğ¿2_ğ‘™ğ‘œğ‘ ğ‘ =(ğ‘¦_ğ‘ğ‘Ÿğ‘’ğ‘‘ âˆ’ ğ‘¦) **2 
    print(t, loss.item())

    # å€’å‚³é: è¨ˆç®—W1èˆ‡W2å°lossçš„å¾®åˆ†(æ¢¯åº¦)
    ###<your code>###
    y_pred_grad = 2. * (y_pred - y)
    W2_grad = h_relu.T.mm(y_pred_grad)
    
    h_grad = y_pred_grad.mm(W2.T) * (h > 0.)
    W1_grad = x.T.mm(h_grad)
    
    # åƒæ•¸æ›´æ–°
    ###<your code>###
    W1.data -= learning_rate * W1_grad
    W2.data -= learning_rate * W2_grad
```
